```python
import time
print(time.time())
def is_last_x_days(epoch_time, x_days=1):
    if time.time()-epoch_time<=60*60*24*x_days:
        return True
    else: 
        return False
```

```python
def get_n():
    df = load_df()
    u = interaction.load_user_database()
    o = {}
    users = list(set(df["user_id"]))
    for user_id in users:
        print(user_id)
        posts = list(set(df.query("user_id==@user_id")["post_id"]))
        comments = list(set(df.query("user_id==@user_id")["comment_id"]))
        q = "user_id!=@user_id and\
        ((post_id==@posts and action!='view_post') or\
        (action=='like_comment_post' and comment_id==@comments))"
        n = df.query(q)
        o[user_id]=[]
        for i in range(len(n)):
            epoch_time = n.iloc[i]["time"]
            if is_last_x_days(epoch_time,3):
                user = n.iloc[i]["user_id"]
                o[user_id].append({"time": epoch_time,
                           "url_info": get_url_info(n.iloc[i]["post_id"]),
                           "users":  u.query("id==@user")["name"].iloc[0],
                           "user_photo_url": get_user_photo_url(n.iloc[i]["user_id"]),
                           "action": n.iloc[i]["action"]})
    print(o)
    with open("notifications.txt", "w") as f:
        f.write(json.dumps(o))
s=time.time()
get_n()
print(time.time()-s)
```
